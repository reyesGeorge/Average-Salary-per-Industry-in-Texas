## Industry Data for 2017
Try and establish standard metrics to establish a growth timeseries per industry. The purpose is to find growing industries who could benefit from our software.

```python
import pandas as pd
import numpy as np
```

```python
iData = pd.read_csv('industryData.csv', thousands=None,header=None, names=["FIPS","Country","zero","Industry", "Employment_Size", "Firms_Number", "Establishment_Number", "Employment","Drop_Me2", "Drop_Me3", "Annual_Payroll", "Drop4"])
df = iData.drop(['zero', 'Drop_Me2', 'Drop_Me3', 'Drop4'], axis=1)
df.index = np.arange( 1, len(df) + 1)

mask = (df != '0').all(1)
df = df[mask]
to_number = ['Employment', 'Annual_Payroll', 'Establishment_Number', 'Firms_Number']
for col in to_number:
        df[col] = df[col].str.replace(',', '').astype(float)
```

```python
#df.info()
#df.Industry.value_counts()
```

```python
df['Annual_Payroll'] = df['Annual_Payroll'].apply(lambda x: x*1000)
df = df.loc[(df.Industry != 'Total') & (df.FIPS == 48)]
df
```

## Seperating Texas Data by Industry

```python
Agriculture_Forestry_Fishing_and_Hunting = df.loc[df.Industry == 'Agriculture, Forestry, Fishing and Hunting'][1::]
Health_Care_and_Social_Assistance = df.loc[df.Industry == 'Health Care and Social Assistance'][1::]
Accommodation_and_Food_Services = df.loc[df.Industry == 'Accommodation and Food Services'][1::]
Professional_Scientific_and_Technical_Services = df.loc[df.Industry == 'Professional, Scientific, and Technical Services'][1::]
Retail_Trade = df.loc[df.Industry == 'Retail Trade'][1::]
Transportation_and_Warehousing = df.loc[df.Industry == 'Transportation and Warehousing'][1::]
Wholesale_Trade = df.loc[df.Industry == 'Wholesale Trade'][1::]
Finance_and_Insurance = df.loc[df.Industry == 'Finance and Insurance'][1::]
Administrative_and_Support_and_Waste_Management_and_Remediation_Services = df.loc[df.Industry == 'Administrative and Support and Waste Management and Remediation Services'][1::]
Other_Services_Except_Public_Administration = df.loc[df.Industry == 'Other Services (except Public Administration)'][1::]
Educational_Services = df.loc[df.Industry == 'Educational Services'][1::]
Real_Estate_and_Rental_and_Leasing = df.loc[df.Industry == 'Real Estate and Rental and Leasing'][1::]
# Construction = df.loc[(df.Industry == 'Construction') & (df.Employment != '0')]
Construction = df.loc[df.Industry == 'Construction'][1::]
Manufacturing = df.loc[df.Industry == 'Manufacturing'][1::]
Information = df.loc[df.Industry == 'Information'][1::]
Management_of_Companies_and_Enterprises = df.loc[df.Industry == 'Management of Companies and Enterprises'][1::]
Arts_Entertainment_and_Recreation = df.loc[df.Industry == 'Arts, Entertainment, and Recreation'][1::]
Mining_Quarrying_and_Oil_and_Gas_Extraction = df.loc[df.Industry == 'Mining, Quarrying, and Oil and Gas Extraction'][1::]
Utilities = df.loc[df.Industry == 'Utilities'][1::]
Industries_not_classified = df.loc[df.Industry == 'Industries not classified'][1::]
```

## EFR = Employee to Firms Average Ratio
## APFR = Annual Payroll to Firms Average Ratio

```python
EFR_BIG_AG = Agriculture_Forestry_Fishing_and_Hunting['Annual_Payroll'].div(Agriculture_Forestry_Fishing_and_Hunting['Employment'])
APFR_BIG_AG = Agriculture_Forestry_Fishing_and_Hunting['Annual_Payroll'].div(Agriculture_Forestry_Fishing_and_Hunting['Firms_Number'])

EFR_Health_n_Social = Health_Care_and_Social_Assistance['Annual_Payroll'].div(Health_Care_and_Social_Assistance['Employment'])
APFR_Health_n_Social = Health_Care_and_Social_Assistance['Annual_Payroll'].div(Health_Care_and_Social_Assistance['Firms_Number'])

EFR_Accommodation_and_Food_Services = Accommodation_and_Food_Services['Annual_Payroll'].div(Accommodation_and_Food_Services['Employment'])
APFR_Accommodation_and_Food_Services = Accommodation_and_Food_Services['Annual_Payroll'].div(Accommodation_and_Food_Services['Firms_Number'])

EFR_Professional_Scientific_and_Technical_Services = Professional_Scientific_and_Technical_Services['Annual_Payroll'].div(Professional_Scientific_and_Technical_Services['Employment'])
APFR_Professional_Scientific_and_Technical_Services = Professional_Scientific_and_Technical_Services['Annual_Payroll'].div(Professional_Scientific_and_Technical_Services['Firms_Number'])

EFR_Retail_Trade = Retail_Trade['Annual_Payroll'].div(Retail_Trade['Employment'])
APFR_Retail_Trade = Retail_Trade['Annual_Payroll'].div(Retail_Trade['Firms_Number'])

EFR_Transportation_and_Warehousing = Transportation_and_Warehousing['Annual_Payroll'].div(Transportation_and_Warehousing['Employment'])
APFR_Transportation_and_Warehousing = Transportation_and_Warehousing['Annual_Payroll'].div(Transportation_and_Warehousing['Firms_Number'])

EFR_Wholesale_Trade = Wholesale_Trade['Annual_Payroll'].div(Wholesale_Trade['Employment'])
APFR_Wholesale_Trade = Wholesale_Trade['Annual_Payroll'].div(Wholesale_Trade['Firms_Number'])

EFR_Finance_and_Insurance = Finance_and_Insurance['Annual_Payroll'].div(Finance_and_Insurance['Employment'])
APFR_Finance_and_Insurance = Finance_and_Insurance['Annual_Payroll'].div(Finance_and_Insurance['Firms_Number'])

EFR_Administrative_and_Support_and_Waste_Management_and_Remediation_Services = Administrative_and_Support_and_Waste_Management_and_Remediation_Services['Annual_Payroll'].div(Administrative_and_Support_and_Waste_Management_and_Remediation_Services['Employment'])
APFR_Administrative_and_Support_and_Waste_Management_and_Remediation_Services = Administrative_and_Support_and_Waste_Management_and_Remediation_Services['Annual_Payroll'].div(Administrative_and_Support_and_Waste_Management_and_Remediation_Services['Firms_Number'])

EFR_Other_Services_Except_Public_Administration = Other_Services_Except_Public_Administration['Annual_Payroll'].div(Other_Services_Except_Public_Administration['Employment'])
APFR_Other_Services_Except_Public_Administration = Other_Services_Except_Public_Administration['Annual_Payroll'].div(Other_Services_Except_Public_Administration['Firms_Number'])

EFR_Educational_Services = Educational_Services['Annual_Payroll'].div(Educational_Services['Employment'])
APFR_Educational_Services = Educational_Services['Annual_Payroll'].div(Educational_Services['Firms_Number'])

EFR_Real_Estate_and_Rental_and_Leasing = Real_Estate_and_Rental_and_Leasing['Annual_Payroll'].div(Real_Estate_and_Rental_and_Leasing['Employment'])
APFR_Real_Estate_and_Rental_and_Leasing = Real_Estate_and_Rental_and_Leasing['Annual_Payroll'].div(Real_Estate_and_Rental_and_Leasing['Firms_Number'])

EFR_Construction = Construction['Annual_Payroll'].div(Construction['Employment'])
APFR_Construction = Construction['Annual_Payroll'].div(Construction['Firms_Number'])

EFR_Manufacturing = Manufacturing['Annual_Payroll'].div(Manufacturing['Employment'])
APFR_Manufacturing = Manufacturing['Annual_Payroll'].div(Manufacturing['Firms_Number'])

EFR_Information = Information['Annual_Payroll'].div(Information['Employment'])
APFR_Information = Information['Annual_Payroll'].div(Information['Firms_Number'])

EFR_Management_of_Companies_and_Enterprises = Management_of_Companies_and_Enterprises['Annual_Payroll'].div(Management_of_Companies_and_Enterprises['Employment'])
APFR_Management_of_Companies_and_Enterprises = Management_of_Companies_and_Enterprises['Annual_Payroll'].div(Management_of_Companies_and_Enterprises['Firms_Number'])

EFR_Arts_Entertainment_and_Recreation = Arts_Entertainment_and_Recreation['Annual_Payroll'].div(Arts_Entertainment_and_Recreation['Employment'])
APFR_Arts_Entertainment_and_Recreation = Arts_Entertainment_and_Recreation['Annual_Payroll'].div(Arts_Entertainment_and_Recreation['Firms_Number'])

EFR_Mining_Quarrying_and_Oil_and_Gas_Extraction = Mining_Quarrying_and_Oil_and_Gas_Extraction['Annual_Payroll'].div(Mining_Quarrying_and_Oil_and_Gas_Extraction['Employment'])
APFR_Mining_Quarrying_and_Oil_and_Gas_Extraction = Mining_Quarrying_and_Oil_and_Gas_Extraction['Annual_Payroll'].div(Mining_Quarrying_and_Oil_and_Gas_Extraction['Firms_Number'])

EFR_Utilities = Utilities['Annual_Payroll'].div(Utilities['Employment'])
APFR_Utilities = Utilities['Annual_Payroll'].div(Utilities['Firms_Number'])

EFR_Industries_not_classified = Industries_not_classified['Annual_Payroll'].div(Industries_not_classified['Employment'])
APFR_Industries_not_classified = Industries_not_classified['Annual_Payroll'].div(Industries_not_classified['Firms_Number'])
```
